```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/Traffic_Collision_Data_from_2010_to_Present.csv', delimiter=',')
```

```python
# Parse coordinates into distinct variables
str_coord = []
for i in range(len(df['Location'])):
    temp_str = df['Location'][i].replace('(', '').replace(')', '').replace(',', '').split(' ')
    temp_coord = [float(x) for x in temp_str]
    str_coord.append(temp_coord)
```

```python
result = []

for i in range(len(str_coord)):
    result.append({
      "type": "Feature",
      "geometry": {
        "type": "Point",
        "coordinates": str_coord[i]
      },
      "properties": {
        "id": int(df['DR Number'][i]),
        "Date Reported": df['Date Reported'][i],
        "Date Occurred": df['Date Occurred'][i]
        }
    })
```

```python
result
```

```python
full_json = {
"type": "FeatureCollection",
"features": result
}
```

```python
from geojson import dump
with open('../data/traffic.geojson', 'w') as out:
    dump(full_json, out)
```
