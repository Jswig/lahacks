```python
import pandas as pd

df = pd.read_csv('../data/Updated_Traffic_Weights.csv', delimiter=',')
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
# Parse coordinates into distinct variables
str_coord = []
for i in range(len(df['location'])):
    temp_str = df['location'][i].replace('(', '').replace(')', '').replace(',', '').split(' ')
    temp_coord = [float(x) for x in reversed(temp_str)]
    str_coord.append(temp_coord)
```

```python
result = []

for i in range(len(str_coord)):
    result.append({
      "type": "Feature",
      "geometry": {
        "type": "Point",
        "coordinates": str_coord[i]
      },
      "properties": {
        "num_accidents": int(df['number_of_accidents'][i])
        }
    })
```

```python
full_json = {
"type": "FeatureCollection",
"features": result
}
```

```python
from geojson import dump
with open('../data/num_accidents.geojson', 'w') as out:
    dump(full_json, out)
```
